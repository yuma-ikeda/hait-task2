```python
#モジュールのインポート
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#Excelファイルの読み込み
Data_train = pd.read_csv('Downloads/train.csv')
print(Data_train)
```

```python
#欠損値をカウント
Data_train.isnull().sum()
```

```python
#ダミー変数作成
pd.get_dummies(Data_train['Sex'])
```

```python
#ダミー変数作成
pd.get_dummies(Data_train['Embarked'])
```

```python
#列の削除
del Data_train['Name']
del Data_train['Ticket']
del Data_train['Cabin']
```

```python
#ダミー変数作成、表示
Data_train['Sex']=pd.get_dummies(Data_train['Sex'])
Data_train['Embarked']=pd.get_dummies(Data_train['Embarked'])
print(Data_train.head(20))
```

```python
#平均値補完
df = Data_train.fillna(Data_train.mean())
print(df)
```

```python
#要素の抽出
X = df.iloc[:,2:8].values
y = df.iloc[:,1].values
```

```python
#ホールドアウト法の実行
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
#調整したいパラメタを指定
param_grid = {'C':[0.1,1.0,10, 100, 1000, 10000],
                        'gamma': [0.001, 0.01, 0.1, 1, 10]}
```

```python
#データの分割方法を指定
from sklearn.model_selection import StratifiedKFold
kf_5 = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
```

```python
# GridSearchCVのインスタンスを生成
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
gs_svc = GridSearchCV(SVC(), param_grid, cv=kf_5)
```

```python
#学習実行
gs_svc.fit(X_train, y_train)
```

```python
# test精度の平均が最も高かった組み合わせを出力
gs_svc.best_params_
```

```python
# そのときのtest精度の平均を出力
gs_svc.best_score_
```

```python
# ホールド・アウト法によるtest精度を出力
gs_svc.score(X_test, y_test)
```
